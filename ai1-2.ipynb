```python
import pyfcctab as fcc
import astropy.units as units
from termcolor import colored
```

```python
tables = fcc.load()
# tables = fcc.read()
```

```python
ai_ranges = [
    [3.3*units.GHz, 3.4*units.GHz],
    [3.6*units.GHz, 3.8*units.GHz],
    [6.425*units.GHz, 7.025*units.GHz],
    [7.025*units.GHz, 7.125*units.GHz],
    [10.0*units.GHz, 10.5*units.GHz],
]
```

```python
for ai in ai_ranges:
    print (colored(f"================================ {ai[0]}-{ai[1]}", attrs=["bold"]))
    bands = tables.itu.get_bands(ai[0], ai[1], adjacent=True)
    for b in bands:
        print(f"----------- {b.range_str()}")
        print(b.to_str(highlight_allocations=True))
```

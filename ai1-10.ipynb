```python
import pyfcctab as fcc
import astropy.units as units
from termcolor import colored
```

```python
tables = fcc.load()
```

```python
ai_ranges = [
    [11.7*units.GHz, 12.7*units.GHz],
    [18.1*units.GHz, 18.6*units.GHz],
    [18.8*units.GHz, 20.2*units.GHz],
    [27.5*units.GHz, 30.0*units.GHz],
]
```

```python
for ai in ai_ranges:
    print (colored(f"================================ {ai[0]}-{ai[1]}", attrs=["bold"]))
    bands = tables.itu.get_bands(ai[0], ai[1], adjacent=True)
    for b in bands:
        print(f"----------- {b.range_str()}")
        print(b.to_str(highlight_allocations=True))
```
